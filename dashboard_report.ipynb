### **Objective**
This project aimed to develop an interactive dashboard using data stored in MongoDB, offering valuable insights into the gaming industry. The dashboard provides visualizations of key metrics such as game pricing, memory requirements, developer influence, and other significant factors.

### **Dataset Overview**
The dataset, sourced from Steam, a leading gaming platform, contains comprehensive details on games, including pricing, system requirements, popularity metrics, and developer data.

#### **1. Dataset Features**
##### **Game Details**
- **Game Name:** Title of the game.
- **Genre:** Category or type of game (e.g., Action, Adventure, Simulation).
- **Release Date:** The year and date a game was launched on Steam.

##### **Pricing Information**
- **Maximum Price:** The highest price for a game.
- **Average Price:** The mean price across the dataset.
- **Yearly Pricing Trends:** A historical analysis of pricing fluctuations over time.

##### **System Requirements**
- **Minimum RAM:** The least amount of memory required to run the game, categorized by operating systems:
  - Windows
  - Linux
  - MacOS
- **Recommended RAM:** The suggested memory for optimal gameplay, categorized by OS:
  - Windows
  - Linux
  - MacOS

##### **Popularity Metrics**
- **Popular Tags:** Metadata describing game features (e.g., "Multiplayer," "Singleplayer").
- **PEGI Ratings:** Official classifications from the Pan-European Game Information system, categorizing games based on age-appropriateness.
- **Popularity Word Cloud:** A visual representation of commonly associated tags and genres.

##### **Developer Information**
- **Developers:** Companies or studios responsible for game creation.
- **Famous Developers:** Highlighting top studios based on game output and popularity.

##### **Platform-Specific Details**
- **System Compatibility:** Information on game compatibility with Windows, Linux, and MacOS.

#### **2. PEGI Ratings and Their Importance**
The dataset includes PEGI ratings, which help in:
- **Regulatory Compliance:** Ensuring games align with industry standards.
- **Parental Guidance:** Assisting parents in making informed gaming choices.
- **Market Analysis:** Identifying audience segments based on content ratings.

##### **Key PEGI Classifications in the Dataset**
- **PEGI 3:** Suitable for all ages.
- **PEGI 7:** Includes mild violence or potentially scary scenes.
- **PEGI 12:** Features moderate violence or mild suggestive content.
- **PEGI 16:** Contains realistic violence, drug use, or mature themes.
- **PEGI 18:** Features explicit violence, adult content, and strong language.

These ratings allow for trend analysis in game development and marketing strategies.

#### **3. Key Dataset Statistics**
- **Total Games Analyzed:** 21,000 unique game entries.
- **Pricing Insights:**
  - **Minimum Price:** $0 (free-to-play games)
  - **Maximum Price:** $999 (premium editions, collector bundles)
  - **Average Price:** $445.11
- **System Requirements:**
  - Minimum RAM varies from 2 GB to 8 GB, depending on the OS.
  - Recommended RAM is typically 8 GB for Windows and MacOS, 4 GB for Linux.
